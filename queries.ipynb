```python
%%sql 
DROP VIEW IF EXISTS counts;
CREATE VIEW counts AS
SELECT a.articleID AS articleID, COUNT(c.articleID) AS count
FROM
    Articles a LEFT JOIN
    Comments c ON a.articleID = c.articleID
GROUP BY a.articleID;

```

```python
%%sql

SELECT DISTINCT newDesk,
AVG(headlinepolarity) AS hp,
AVG(headlinesubjectivity) AS hs,
AVG(snippetpolarity) AS sp,
AVG(snippetsubjectivity) AS ss, 
AVG(snippetentropy) AS e
FROM Articles
GROUP BY newDesk
ORDER BY e
```

```python
%%sql
DROP VIEW IF EXISTS Commentcounts;
DROP VIEW IF EXISTS ac;

CREATE VIEW CommentCounts AS
    SELECT c.articleID, cc.count AS Comments_count, c.newDesk AS category,
    AVG(c.polarity) AS Polarity, AVG(c.subjectivity) AS Subjectivity, AVG(c.entropy) AS Entropy,
    AVG(c.replyCount) AS replyCount, a.headlinesubjectivity AS hSubjectivity, a.headlinePolarity AS hPolarity,
    a.snippetsubjectivity AS ssubjectivity, a.snippetpolarity AS spolarity, a.snippetentropy AS sEntropy,
    a.articlewordcount AS awords
    FROM Articles AS a
    LEFT JOIN Comments AS c ON c.articleID = a.articleID
    JOIN counts AS cc ON cc.articleID = a.articleID
    GROUP BY c.articleID, cc.count, category, a.headlinesubjectivity, a.headlinepolarity, a.snippetsubjectivity, a.snippetpolarity, a.snippetentropy, a.articlewordcount;
    
```

```python
%%sql
CREATE VIEW ac AS 
    SELECT a.SnippetPolarity, a.SnippetSubjectivity, a.newDesk, c.comments_count
    FROM Articles AS a
    JOIN CommentCounts AS c ON a.articleID = c.articleID;
    
SELECT newdesk AS Topic, AVG(comments_count) AS AverageComments
FROM ac
GROUP BY newdesk
ORDER BY AverageComments
```
