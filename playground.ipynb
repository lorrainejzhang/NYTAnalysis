```python
def findHeadlinePolarity(row):
    blob = TextBlob(row['headline'])
    return blob.sentiment.polarity

def findHeadlineSubjectivity(row):
    blob = TextBlob(row['headline'])
    return blob.sentiment.subjectivity

def findSnippetPolarity(row):
    blob = TextBlob(row['snippet'])
    return blob.sentiment.polarity

def findSnippetSubjectivity(row):
    blob = TextBlob(row['snippet'])
    return blob.sentiment.subjectivity

def readSnippet(row):
    text = TextBlob(row['snippet'])
    words = text.split(' ')
    counter = dict(collections.Counter(words))
    return counter

def findSnippetEntropy(row):
    counter = readSnippet(row)
    counts = np.array(list(counter.values()))
    entropy = scipy.stats.entropy(counts, base=2)
    return entropy
```

```python
def findPolarity(row):
    blob = TextBlob(row['commentBody'])
    return blob.sentiment.polarity

def findSubjectivity(row):
    blob = TextBlob(row['commentBody'])
    return blob.sentiment.subjectivity

def readSnippet(row):
    text = TextBlob(row['commentBody'])
    words = text.split(' ')
    counter = dict(collections.Counter(words))
    return counter

def findEntropy(row):
    counter = readSnippet(row)
    counts = np.array(list(counter.values()))
    entropy = scipy.stats.entropy(counts, base=2)
    return entropy
```

```python
def findCounter(text):
    text = re.sub(r'\W+', ' ',text)
    words = text.split(' ')
    counter = dict(collections.Counter(words))
    return counter

def findEntropy(text):
    counter = findCounter(text)
    counts = np.array(list(counter.values()))
    entropy = scipy.stats.entropy(counts, base=2)
    return entropy

# def findEntropiesWithCounter(counter):
#     entropies = dict()
#     for k in snippets.keys():
#         entropies[k] = findEntropy(str(snippets[k]))
#     return entropies
#     counts = np.array(list(counter.values()))
#     entropy = scipy.stats.entropy(counts, base=2)
#     return entropy

def findEntropies(snippets):
    entropies = dict()
    for k in snippets.keys():
        entropies[k] = findEntropy(str(snippets[k]))
    return entropies

findEntropies(snippets)
```
